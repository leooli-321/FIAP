## 3. Limpando os dados

1. remover dados faltantes
2. remover duplicatas
3. remover colunas com baixa utilidade

```python
df.info()
```

```python
df = df.dropna()  # Elimina todas as linhas com dados faltantes (pode ser legal armazenar o clean em outro DF)
df = df.fillna(0) # Preenche com zero   as linhas que estão faltando

df.dropna(inplace=True) # Nâo é boa prática pois pode perder dados.
```

```python
df.info() # Mostra  informações do DataFrame
```

```python
duplicadas = df[df.duplicated()] # Duplica  as linhas que estão repetidas

df = df.drop_duplicates() # apaga as linhas duplicadas

print(duplicadas)
```

```python
df = df.drop("user_id", axis=1)
```
